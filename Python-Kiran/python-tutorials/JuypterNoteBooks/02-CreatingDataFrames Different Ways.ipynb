# Method 1: Using Numpy array and Pandas

```python
import numpy as np
import pandas as pd
a=np.arange(25).reshape(5,5)
```

```python
dt=pd.DataFrame(a,columns=['dept1','dept2','dept3','dept4','dept5'])
```

```python
dt
```
