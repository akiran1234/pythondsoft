#  groupby() ==Split data into Groups-->Apply Stat Function --> Combine All                                                                                                                    Results

```python
gd=dt.groupby('Branch')           # Here gd is groupby object.
gd
```

```python
for branch,branch_data in gd:
    print(branch)
    print(branch_data)
```

```python
gd.max()['Total Marks']          # Applying stat functions on group by object. Similarly gd.min(), gd.sum(),gd.count() etc ..
```

```python
gd.describe()['Total Marks']     # describe() will give all the statistics in one shot
```

# concat() will combine the dataframes vertically/horizontally.

```python
df1=pd.read_csv("C:\\Users\\PBR\\Desktop\\Datasets\\Pandas-Datasets\\combine-1.csv")
df2=pd.read_csv("C:\\Users\\PBR\\Desktop\\Datasets\\Pandas-Datasets\\combine-2.csv")
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis=0,keys=['data-1','data-2'])  # concat() will combine two dataframes vertically and horizontally
                                                      # axis=0 combine vertically and axis=1 combine horizontally
```

```python
pd.concat([df1,df2],axis=1,keys=['data-1','data-2'])  # concatinating horizontally and parameters are passed as lists.
 
```

```python
pd.merge(df1,df2,how='inner',on='deptno')
```
