```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips').sort_values('tip',ascending=True)
```

```python
tips.head()
```

```python
sns.stripplot(x='time',y='tip',data=tips,hue='sex',jitter=0.1,dodge=True)
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',split=True,inner='quartile')
```

```python
sns.rugplot(tips['total_bill'],height=0.5)
```

```python
sns.distplot(tips['total_bill'],bins=10,hist=True,rug=True,norm_hist=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips,hue='sex',kind='scatter')
```
